```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
names = ['year', 'month', 'day', 'hour', 'minute', 'second', 'longitude', 'latitude', 'magnitude', 'depth', 'id', 't']
dateparse = lambda y,m,d,h,mn,sec: pd.datetime.strptime(' '.join(map(str,(y,m,d,h,mn,sec))), '%Y %m %d %H %M %S.%f')
df = pd.read_csv('event_all.cat', sep='\s+', names=names, date_parser = dateparse, parse_dates={'dt':['year', 'month', 'day', 'hour', 'minute', 'second']})
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(df.dt, df.depth, marker='o', color='blue', linestyle='None')
ax.set_ylim([0, 15])
plt.gca().invert_yaxis()
```

Histograms of clear and steamy images
--

What are the brightness histograms of clear and steamy histograms and what metrics can we find that may permit use to distinguish between the images.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#clear = '/home/ruma/geonet/volcam_images/histogram/20171214003002WIWR.jpg'
#steam = '/home/ruma/geonet/volcam_images/histogram/20171201000002WIWR.jpg'
clear = '/home/sherburn/geonet/volcam_images/crop/20180101001002WIWR.jpg'
steam = '/home/sherburn/geonet/volcam_images/crop/20180101005002WIWR.jpg'
```

```python
imgclear = cv2.imread(clear)
imgsteam = cv2.imread(steam)
```

**show images**

```python
plt.imshow(imgclear)
```

```python
plt.imshow(imgsteam)
```

**histograms**

```python
histclear = cv2.calcHist([imgclear],[0],None,[256],[0,256])
histsteam = cv2.calcHist([imgsteam],[0],None,[256],[0,256])

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(histclear, color='red', label='clear')
ax.plot(histsteam, color='blue', label='steam')
ax.set_ylabel('number pixels')
ax.set_xlabel('pixel intensity')
ax.set_ylim(ymin=0)
plt.legend(loc='best')
```

**histogram width**

```python
def hist_wid(hist):
    minnonz = np.flatnonzero(hist).min()
    maxnonz = np.flatnonzero(hist).max()
    return (maxnonz - minnonz)
```

```python
widclear = hist_wid(histclear)
widsteam = hist_wid(histsteam)
print ('width clear:', widclear)
print ('width steam:', widsteam)
```

**histogram standard deviation**

```python
def hist_std(hist):
    std = np.std(np.flatnonzero(hist))
    return (std)
```

```python
stdclear = hist_std(histclear)
stdsteam = hist_std(histsteam)
print ('std clear:', stdclear)
print ('std steam:', stdsteam)
```

**histogram maximum**

```python
def hist_max(hist):
    max = np.argmax(hist)
    return (max)
```

```python
maxclear = hist_max(histclear)
maxsteam = hist_max(histsteam)
print ('max clear:', maxclear)
print ('max steam:', maxsteam)
```

Calculate a dataset for steamy image analysis
--

This produces data to be used as a test (or application) of a previously derived model.

Do not enhance contrast.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd

%matplotlib inline
```

```python
def hist_wid(hist):
    minnonz = np.flatnonzero(hist).min()
    maxnonz = np.flatnonzero(hist).max()
    return (maxnonz - minnonz)
```

```python
def hist_std(hist):
    std = np.std(np.flatnonzero(hist))
    return (std)
```

```python
def hist_max(hist):
    max = np.argmax(hist)
    return (max)
```

```python
def imfdtstr(imfile):
    #gets the datetime of an image file named like 20180101001002WIWR.jpg as a string, for labelling only
    dt = imfile[:14]
    return (dt)
```

```python
imagedir  = '/home/sherburn/geonet/volcam_images/crop'
```

```python
images = pd.DataFrame()

for file in sorted(os.listdir(imagedir)):
        if file.endswith('.jpg'):
            #print (file)
            imfile = os.path.join(imagedir, file)
            img = cv2.imread(imfile)
            imtime = imfdtstr(file)
            #calculate histogram
            hist = cv2.calcHist([img],[0],None,[256],[0,256])
            wid = hist_wid(hist) #width
            #std = hist_std(hist) #standard deviation
            #hmax = hist_max(hist) #index of maximum
            d = {'wid':wid}
            df = pd.DataFrame(d, index=[imtime])
            images = images.append(df)

images.to_csv('new_image_data.csv', columns=['wid'])            
```

# Using Python to Examine the LaMEVE Volcanic Eruption Database

How to get some useful information from the Large Magnitude Explosive Volcanic Eruptions (LaMEVE) Database using basic python. Get the database, as a tab-deliminted text file from http://www.bgs.ac.uk/vogripa/searchVOGRIPA.cfc?method=searchForm.
The file has 7735 entries and 90 columns = 696,000 cells.

```python
import pandas as pd
%matplotlib inline
#pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier

# This is necessary to show lots of columns in pandas 0.12. 
# Not necessary in pandas 0.13.
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 100)
```

Read the database file and find some basic information
---

```python
df = pd.read_csv('lameve.txt', sep='\t', encoding='latin1')

print len(df)
df.head()
```

Find how many regions and sub-regions are represented?

```python
regions = df['Region'].value_counts().count()
print regions
```

```python
subregions = df['Sub-region'].value_counts().count()
print subregions
```

Show eruptions in each region

```python
df['Region'].value_counts().plot(kind='barh')
```

```python
df['Region'].value_counts().plot.pie(figsize=(6, 6))
```

## Okataina and Taupo eruptions

```python
okataina = nz[nz['Volcano Name']=='Okataina']
taupo = nz[nz['Volcano Name']=='Taupo']
```

Plot date of VEI of eruptions from both volcanoes on the same axis

```python
ax1 = okataina.plot.scatter(x='Eruption start year', y='VEI', color = 'red', label = 'Okataina')
taupo.plot.scatter(x='Eruption start year', y='VEI', color='blue', label='Taupo', ax=ax1)
```

Eruption Magnitude

```python
ax1 = okataina.plot.scatter(x='Eruption start year', y='Magnitude', color = 'red', label = 'Okataina')
taupo.plot.scatter(x='Eruption start year', y='Magnitude', color='blue', label='Taupo', ax=ax1)
```

Maximum column height vs VEI, a big plot

```python
ax1 = okataina.plot.scatter(x='VEI', y='Column height Max', color = 'red', label = 'Okataina',figsize=(15, 8))
taupo.plot.scatter(x='VEI', y='Column height Max', color='blue', label='Taupo', ax=ax1)
```

Re-read database

```python
df = pd.read_csv('lameve.txt', sep='\t', encoding='latin1')
df.head()
```

Dump columns that contain 'Citation',which is causing the duplic

```python
import re
df2 = df.select(lambda x: not re.match('Citation', x), axis=1)
```

```python
df2.drop_duplicates(inplace=True)
len(df2)
```

```python
nz = df2[df2['Sub-region']=='New Zealand']
len(nz)
```

Re-do some things - Results are now quite different without duplicates
